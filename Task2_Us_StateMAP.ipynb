### bar chart

```python
try:
    ggplot(us_state_df, aes(x='State', fill='Company1')) + geom_bar(stat='count')
    ggplot(us_state_df, aes(x='State', fill='Company2')) + geom_bar(stat='count')
    plt1 = ggplot(us_state_df, aes(x='State', fill='Company1')) + geom_bar(stat='count')
    plt1.save(filename="plot1.png", plot=plt1, device='png', dpi=300, height=25, width=25)
    plt2 = ggplot(us_state_df, aes(x='State', fill='Company2')) + geom_bar(stat='count')
    plt2.save(filename="plot2.png", plot=plt2, device='png', dpi=300, height=25, width=25)
except Exception as ex:
    exc_type, exc_obj, tb = sys.exc_info()
    template = "Main: An exception of type {0} occurred at {1}. Arguments:\n{2!r}"
    message = template.format(type(ex).__name__, tb.tb_lineno, ex.args)
    logging.error(message)
```
